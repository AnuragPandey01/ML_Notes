# Types of Machine Learning

## 1. **Supervised Learning**  
In supervised learning, the model is trained on labeled data, meaning each training example is paired with an output label.  

### a. Classification  
   - Used when the output variable is categorical.  
   - **Examples:**  
     - Decision Tree Classification  
     - Random Forest Classification  
     - K-Nearest Neighbors (KNN)  
     - Support Vector Machine (SVM)  
     - Logistic Regression  

### b. Regression  
   - Used when the output variable is continuous.  
   - **Examples:**  
     - Linear Regression  
     - Polynomial Regression  
     - Ridge Regression  
     - Lasso Regression  
     - Support Vector Regression (SVR)  
     - Neural Networks (for regression tasks)  

## 2. **Unsupervised Learning**  
Unsupervised learning deals with unlabeled data, aiming to find patterns or groupings within the data.  

### a. Clustering  
   - Used to group similar data points together.  
   - **Examples:**  
     - K-Means Clustering  
     - Hierarchical Clustering  
     - DBSCAN (Density-Based Spatial Clustering)  
     - Gaussian Mixture Models (GMM)  

### b. Dimensionality Reduction  
   - Used to reduce the number of input features while preserving important information.  
   - **Examples:**  
     - Principal Component Analysis (PCA)  
     - t-Distributed Stochastic Neighbor Embedding (t-SNE)  
     - Autoencoders  

### c. Association Rule Learning  
   - Used to discover relationships between variables in large datasets.  
   - **Examples:**  
     - Apriori Algorithm  
     - Eclat Algorithm  

## 3. **Reinforcement Learning**  
Reinforcement learning (RL) involves an agent that learns to make decisions by interacting with an environment to maximize cumulative rewards.  

### a. Model-Free RL  
   - The agent learns directly from experiences without an explicit model of the environment.  
   - **Examples:**  
     - Q-Learning  
     - Deep Q Networks (DQN)  
     - SARSA  

### b. Model-Based RL  
   - The agent builds a model of the environment to plan actions.  
   - **Examples:**  
     - AlphaGo  
     - Monte Carlo Tree Search  

### c. Policy-Based and Value-Based Methods  
   - **Policy-Based:** Directly optimizes the policy (e.g., REINFORCE, PPO).  
   - **Value-Based:** Uses value functions to determine the best action (e.g., Q-Learning, DDPG).
